```python
import math 

def karatsuba(x,y):
    
    if (x<10 or y<10):
     return x*y
    
    n = max(len(str(x)),len(str(y)))
    n1 = n//2
 
            
    a = x //(10**n1)
    b = x %(10**n1)
    c = y //(10**n1)
    d = y %(10**n1)
    
    
    
    ac = karatsuba(a,c)
    bd = karatsuba(b,d)
    
    G = karatsuba(a+b, c+d) - ac - bd
    
   
    P = (10**(2*n1))*ac + (10**(n1)) * G + bd
    
    return P
    

    
```
